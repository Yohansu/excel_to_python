```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
```

```python
print("Numpy version is: ", np.__version__)
print("Pandas version is: ", pd.__version__)
```

```python
# Reading the data
df_raw = pd.read_excel("raw_data.xlsx")
```

```python
# Filtering dates from 2010 to 2025

# Create a new dataframe that includes Date, Open and Close columns
df_clean = pd.DataFrame(df_raw, columns=['Date', 'Open' ,'Close'])
#df_clean.head()
```

```python
# Enforce datetime format to Date column
df_clean["Date"] = pd.to_datetime(df_clean["Date"], errors="coerce")

df_clean= df_clean[df_clean["Date"].between("2010-01-01", "2025-12-31")]
#df_clean["Date"].min(), df_clean["Date"].max() 
```

```python
# Look up for closing prices in certain dates - second version

dates = ["2025-12-26", "2024-01-02"]

closing_prices = {
    d: df_clean.loc[df_clean["Date"] == d, "Close"].iloc[0]
    for d in dates
    if (df_clean["Date"] == d).any()
}

#closing_prices

# Print two specific dates closing prices

print(f"{dates[0]}: Closing price = {closing_prices[dates[0]]}")
print(f"{dates[1]}: Closing price = {closing_prices[dates[1]]}")
```

```python
# Computing day type: if the opening price is greater than the closing price then its a "Down day" and if its not then its an "Up day"

# Create a new column called "Day_type
# Using np.where allow us to look for a given condition and telling it where to look for, it returns an array that satisfy or not the given condition
df_clean["Day_type"] = np.where(
    df_clean["Close"] > df_clean["Open"], "Up",
    np.where(df_clean["Close"] < df_clean["Open"], "Down", "Flat")
)
```

```python
df_clean
```

```python
# Counting Up days and Down days
#df_clean["Day_type"].value_counts()

print(f"Count Up days = {df_clean["Day_type"].value_counts()["Up"]}")
print(f"Count Down days = {df_clean["Day_type"].value_counts()["Down"]}")
```

```python
# Compute the average of all closing prices that had a Up day
avg_close_by_type = df_clean.loc[df_clean["Day_type"] == "Up", "Close"].mean()
avg_close_down = df_clean.loc[df_clean["Day_type"] == "Down", "Close"].mean()

print(f"Average closing price for all Up days: {round(avg_close_by_type,)}")
print(f"Average closing price for all Down days: {round(avg_close_down)}")
```

```python
# Compute STD, mean, variance and median for closing prices

std_close = df_clean["Close"].std()  # STD
mean_close = df_clean["Close"].mean() # mean
variance_close = std_close ** 2       # variance
median_close = df_clean["Close"].median() # median


print(f"The STD for the closing prices is: {round(std_close)}")
print(f"The mean for the closing prices is: {round(mean_close)}")
print(f"The variance for the closing prices is: {round(variance_close)}")
print(f"The median for the closing prices is: {round(median_close)}")
```

```python
# Plot S&P 500 Closing prices over time, from 2010 to 2025
plt.figure(figsize=(10,5))
plt.plot(df_clean["Date"] ,df_clean["Close"])
plt.xlabel("Date")
plt.ylabel("Close price")
plt.title("S&P 500 Closing prices over time, from 2010 to 2025")
plt.show()
```

```python
# Plot the distribution of S&P 500 closing prices
plt.figure(figsize=(10,5))
plt.hist(df_clean["Close"], bins=50)
plt.xlabel("Closing prices")
plt.ylabel("Frequency")
plt.title("Distribution of S&P 500 closing prices")
plt.show()
```
